```python
import pandas as pd
import numpy as np
df=pd.read_csv("/home/xinyang/Kaggle/TRD/TRD_Dalyr.csv",encoding='utf-16',sep='\t')
df = df.reset_index()
df
```

```python
# Representing Data with EntitySets
df_target=df[["Dsmvtll","Dsmvosd","Dnvaltrd"]]
df_target.reset_index()
df_var=df.drop(["Dsmvtll","Dsmvosd","Dnvaltrd"],axis=1)
import featuretools as ft
es = ft.EntitySet(id="TRD_data")
#add entities
es = es.entity_from_dataframe(entity_id="TRD_target",
                               dataframe=df_target,
                             index="index")
es = es.entity_from_dataframe(entity_id="TRD_var",
                               dataframe=df_var,
                             index="index")
es
```

```python
feature_matrix, feature_defs = ft.dfs(entityset=es,
                                      target_entity="TRD_target",
                                      agg_primitives=["mean","max","median"],
                                       trans_primitives=["month", "hour"],
                                     max_depth=2)
feature_matrix
```
