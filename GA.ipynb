```python
import pandas as pd
import numpy as np
#years = ['2005','2006','2007','2008','2009','2010','2011','2012','2013','2014','2015','2016','2017','2018','2019']
years = ['2005']
for year in years:
    y = 'con'+year+'.csv'
    df_con =pd.read_csv(y)
    df_con = df_con.drop(df_con.columns[[0]], axis=1)
    df_con = df_con.drop('year',1)
    for i in range(9):
        columns = []
        column_name = list(df_con.columns.values)
        for j in range(17):
            columns.append(str(i)+'_'+column_name[j])
        df = df_con.iloc[(i+1):,0:17].append(df_con.iloc[:(i+1),0:17])
        df = df.reset_index()
        df = df.drop('index',1)
        df.columns= columns
        df_con = pd.concat([df_con,df],axis=1)
    df_merge = pd.DataFrame()
    for i in range(10):
        j =7
        name = str(i)+'_'+str(j)
        df = df_con.iloc[i:,6].append(df_con.iloc[:i,6])
        df = df.reset_index()
        df = df.drop('index',1)
        df_merge[name]=list(df.iloc[:,0])
    df_merge['mean'] = df_merge.mean(axis=1)
    df_merge['std10'] = df_merge.std(axis=1) 
    df_merge['Stkcd'] = df_con['Stkcd']
    df_merge['Trddt'] = df_con['Trddt']
    df1 = df_con['Dretnd']
    df1 = df1.iloc[10:].append(df1.iloc[:10])
    df1 = df1.reset_index()
    df1 = df1.drop('index',1)
    df_merge['np'] = list(df1.iloc[:,0])
    df_con['mean'] = df_merge['mean']
    df_con['std10'] = df_merge['std10']
    df_con['exp'] = np.round(df_merge['np']/df_merge['std10'],8)
    df_con =df_con.sort_values(by=['Trddt'])
    df_con = df_con.reset_index()
    df_con = df_con.drop('index',1)
    date = []
    date_old = df_con.loc[0,'Trddt']
    d = 0
    for i in range(df_con.shape[0]):
        if df_con.loc[i,'Trddt'] == date_old:
            date.append(d)
        else:
            d += 1
            date.append(d)
            date_old = df_con.loc[i,'Trddt']
    df_con['date'] = date
    df_con= df_con[df_con['date']<max(date)-8]
    df_con['goal'] = df_con.groupby('date')['exp'].rank(ascending=0,method='dense')
    df_con = df_con.drop(['exp','date'],1)
    names = ['Adjprcnd','Adjprcwd','Clsprc','Dnshrtrd','Dnvaltrd','Dretnd','Dretwd','Dsmvosd','Dsmvtll','Hiprc','Loprc','Markettype','Opnprc','Stkcd']
    for i in range(14):
        for j in range(10):
            if j == 0:
                name = names[i]
            else:
                name = str(j-1)+'_'+names[i]
            df = df_con[name]
            n = 'cot_'+name
            df_con[n] = 1/np.tan(df)
            n = 'ln_'+name
            df_con[n] = np.log(df)
            df_con[n] = df_con[n].fillna(-10000)
            n = 'bool_'+name
            df_con[n] = df>0
            n = 'abs_'+name
            df_con[n] = np.absolute(df)
            if names[i] =='Dretnd':
                n = 'ceil_'+name
                df_con[n] = np.ceil(df)
                n = 'floor_'+name
                df_con[n] = np.floor(df)
                n = 'round_'+name
                df_con[n] = np.round(df)
    n = 'df_out'+year+'.csv'
    df_out = df_con.sort_values(by=['Stkcd','Trddt'])
    df_out = df_out.reset_index()
    df_out = df_out.drop(['index'],1)
    df_out.to_csv(n)
```

```python
import geppy as gep
from deap import creator, base, tools
import numpy as np
import random
# for reproduction
s = 0
random.seed(s)
np.random.seed(s)
```

```python
def protected_div(x1, x2):
    if abs(x2) < 1e-6:
        return 1
    return x1 / x2
import operator 

pset = gep.PrimitiveSet('Main', input_names=['x'])
pset.add_function(operator.add, 2)
pset.add_function(operator.sub, 2)
pset.add_function(operator.mul, 2)
pset.add_function(protected_div, 2)
pset.add_rnc_terminal()

from deap import creator, base, tools

creator.create("FitnessMin", base.Fitness, weights=(-1,))  # to minimize the objective (fitness)
creator.create("Individual", gep.Chromosome, fitness=creator.FitnessMin)

h = 6 # head length
n_genes = 2   # number of genes in a chromosome
r = 8   # length of the RNC array
toolbox = gep.Toolbox()
toolbox.register('rnc_gen', random.randint, a=-5, b=5)   # each RNC is random integer within [-5, 5]
toolbox.register('gene_gen', gep.GeneDc, pset=pset, head_length=h, rnc_gen=toolbox.rnc_gen, rnc_array_length=r)
toolbox.register('individual', creator.Individual, gene_gen=toolbox.gene_gen, n_genes=n_genes, linker=operator.add)
toolbox.register("population", tools.initRepeat, list, toolbox.individual)

# compile utility: which translates an individual into an executable function (Lambda)
toolbox.register('compile', gep.compile_, pset=pset)
X = df_con.drop(['goal','Capchgdt','Trddt'],axis=1)
X = X[X!=True]
X = X[X!=False]
Y = df_con['goal']
def evaluate(individual):
    """Evalute the fitness of an individual: MAE (mean absolute error)"""
    func = toolbox.compile(individual)
    Yp = np.array(list(map(func, X)))
    return np.mean(np.abs(Y - Yp)),
toolbox.register('evaluate', evaluate)
```

```python
toolbox.register('select', tools.selTournament, tournsize=3)
# 1. general operators
toolbox.register('mut_uniform', gep.mutate_uniform, pset=pset, ind_pb=0.05, pb=1)
toolbox.register('mut_invert', gep.invert, pb=0.1)
toolbox.register('mut_is_transpose', gep.is_transpose, pb=0.1)
toolbox.register('mut_ris_transpose', gep.ris_transpose, pb=0.1)
toolbox.register('mut_gene_transpose', gep.gene_transpose, pb=0.1)
toolbox.register('cx_1p', gep.crossover_one_point, pb=0.3)
toolbox.register('cx_2p', gep.crossover_two_point, pb=0.2)
toolbox.register('cx_gene', gep.crossover_gene, pb=0.1)
# 2. Dc-specific operators
toolbox.register('mut_dc', gep.mutate_uniform_dc, ind_pb=0.05, pb=1)
toolbox.register('mut_invert_dc', gep.invert_dc, pb=0.1)
toolbox.register('mut_transpose_dc', gep.transpose_dc, pb=0.1)
# for some uniform mutations, we can also assign the ind_pb a string to indicate our expected number of point mutations in an individual
toolbox.register('mut_rnc_array_dc', gep.mutate_rnc_array_dc, rnc_gen=toolbox.rnc_gen, ind_pb='0.5p')
toolbox.pbs['mut_rnc_array_dc'] = 1  # we can also give the probability via the pbs property
```

```python
stats = tools.Statistics(key=lambda ind: ind.fitness.values[0])
stats.register("avg", np.mean)
stats.register("std", np.std)
stats.register("min", np.min)
stats.register("max", np.max)
```

```python
# size of population and number of generations
n_pop = 100
n_gen = 100

pop = toolbox.population(n=n_pop)
hof = tools.HallOfFame(3)   # only record the best three individuals ever found in all generations

# start evolution
pop, log = gep.gep_simple(pop, toolbox, n_generations=n_gen, n_elites=1,
                          stats=stats, hall_of_fame=hof, verbose=True)
```
